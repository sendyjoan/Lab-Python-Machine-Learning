```python
import pandas as pd
import matplotlib.pyplot as plt 
import plotly.graph_objs as go
import numpy as np
import statistics

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Statistik Komputasi/Dataset Tugas/data-jumlah-peserta-kb-menurut-metode-kontrasepsi-dki-jakarta-tahun-2019-rev.csv')
print(data)
```

```python
#Tugas Pertemuan 3&4 : Pengumpulan dan Penyajian Data
#Membuat Bar Chart dengan dataset
plt.title("Peserta KB Aktif Provinsi DKI Jakarta Tahun 2019")
plt.ylabel("Jumlah Peserta")
plt.bar(x = data["jenis_kontrasepsi"], height = data["jumlah_peserta_kb_aktif"])
plt.show
```

```python
plt.title("Peserta KB Aktif Provinsi DKI Jakarta Tahun 2019")
plt.ylabel("Jumlah Peserta")
plt.plot(data["jenis_kontrasepsi"], data["jumlah_peserta_kb_aktif"], linestyle='dashed')
plt.show()
```
