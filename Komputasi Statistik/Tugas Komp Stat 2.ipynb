```python
import pandas as pd
import matplotlib.pyplot as plt 
import statistics

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Statistik Komputasi/Dataset Tugas/data-jumlah-vaksinasi-lansia-menurut-kelurahan-di-provinsi-dki-jakarta-bulan-oktober-tahun-2021.csv')
print(data)
```

```python
slicing_1 = data.iloc[0:8277]
slicing_1.head()
```

```python
slicing_2 = data.iloc[8278:]
slicing_2.head()
```

```python
plt.bar(x = slicing_1['kelurahan'], height = slicing_1['jumlah_vaksin'])
plt.show
mean_slicing_1 = statistics.mean(slicing_1['jumlah_vaksin'])
median_slicing_1 = statistics.median(slicing_1['jumlah_vaksin'])
print('Mean Vaksin Dosis 1')
print(mean_slicing_1)
print('Median Vaksin Dosis 1')
print(median_slicing_1)
```

```python
plt.bar(x = slicing_2['kelurahan'], height = slicing_2['jumlah_vaksin'])
plt.show
mean_slicing_2 = statistics.mean(slicing_2['jumlah_vaksin'])
median_slicing_2 = statistics.median(slicing_2['jumlah_vaksin'])
print('Mean Vaksin Dosis 2')
print(mean_slicing_2)
print('Median Vaksin Dosis 2')
print(median_slicing_2)
```

```python
#Tugas Pertemuan 3&4 : Pengumpulan dan Penyajian Data
#Membuat Bar Chart dengan dataset yang sudah digunakan pada pertemuan 2! 
plt.title("Vaksin Lansia Provinsi DKI Jakarta")
plt.ylabel("Jumlah Vaksin Disalurkan")
plt.bar(x = data["dosis"], height = data["jumlah_vaksin"])
plt.show
```

```python
plt.pie(data["jumlah_vaksin"], labels = data["dosis"], shadow = True, colors="rgbcm")
plt.title("Peserta KB Aktif Provinsi DKI Jakarta Tahun 2019")
plt.legend(title="Jenis Kontrasepsi")
plt.show()
```
