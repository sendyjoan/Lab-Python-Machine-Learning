# Persiapan Data

```python
import pandas as pd

data = pd.read_csv('https://storage.googleapis.com/sendyjoan_bucket1/dataset_quiz.csv', sep=';')

data.head()
```

```python
# Melakukan pengecekan jumlah kolom null
data.isna().sum()
```

# Soal No 1
Buat 5 jenis visualisasi yang dapat dihasilkan dari dataset yang diberikan! (5 poin)

```python
# Melakukan Visualisasi 1 (Jumlah Peserta Volunteer yang tergabung pada setiap tahunnya)
import matplotlib.pyplot as plt
%matplotlib inline

vis1 = data.groupby(['tahun']).count().reset_index()
vis1.sort_values('jml_vltr_terlibat', ascending=True, inplace=True)
fig = plt.figure(figsize=(6,4))
ax = plt.axes()
plt.xlabel('Jumlah')
plt.title('Jumlah Volunter Yang Tergabung Berdasarkan Tahun')
ax.barh(vis1['tahun'],vis1['jml_vltr_terlibat'])
plt.xticks(rotation=90)
plt.show()
print('\n')

# Melakukan Visualisasi 2 (Perbandingan Skema Pendanaan Pada Dataset)
vis2 = data.groupby(['skema']).count().reset_index()
vis2.head()
y = vis2['tahun']
mylabels = ['Kemitraan', 'Reguler']
myexplode = [0.2, 0]
plt.figure(figsize=(6,4))
plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode, autopct='%1.1f%%')
plt.title("Perbandingan Skema Pendanaan Pada Dataset \n")
plt.show() 
print('\n')

# Melakukan Visualisasi 3 (Potensi Luaran HKI Pada Dataset)
vis3 = data.groupby(['potensi_hki']).count().reset_index()
vis3.head()
plt.figure(figsize=(6,4))
ax = plt.axes()
plt.ylabel('Jumlah')
plt.title('Potensi Luaran HKI Pada Dataset')
ax.bar(vis3['potensi_hki'],vis3['skema'])
plt.xticks(rotation=90)
plt.show()
print('\n')

# Melakukan Visualisasi 4 (Status Luaran Lain Pada Dataset)
vis4 = data.groupby(['status_luaran_lain']).count().reset_index()
vis4.head()
plt.figure(figsize=(6,4))
ax = plt.axes()
plt.ylabel('Jumlah')
plt.title('Status Luaran Lain Pada Dataset')
ax.bar(vis4['status_luaran_lain'],vis4['skema'])
plt.xticks(rotation=90)
plt.show()
print('\n')

# Melakukan Visualisasi 5 (Perbandingan Keterlibatan Volunter Pada Dataset)
vis5 = data.groupby(['keterlibatan_vltr']).count().reset_index()
vis5.head()
y = vis5['tahun']
mylabels = ['Tidak', 'Ya']
myexplode = [0.2, 0]
plt.figure(figsize=(6,4))
plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode, autopct='%1.1f%%', shadow = True)
plt.title("Perbandingan Keterlibatan Volunter Pada Dataset \n")
plt.show() 
```

```python
# Melakukan Split Data dengan perbandingan data training : validation : testing sebesar 8 : 1 : 1
from sklearn.model_selection import train_test_split

df_train, df_unseen = train_test_split(data, test_size=0.2, random_state=0)
df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0)

print(f'Jumlah data asli: {data.shape[0]}')
print(f'Jumlah data train: {df_train.shape[0]}')
print(f'Jumlah data val: {df_val.shape[0]}')
print(f'Jumlah data test: {df_test.shape[0]}')
```

# Soal No 4
Terapkan dua jenis metode klasifikasi dan lakukan perbandingan akurasi  dengan wajib melibatkan variabel keterlibatan volunteer sampai variabel luaran-luaran yang dihasilkan! (20 poin)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
```

```python
df_train['potensi_hki'] = encode.fit_transform(df_train['potensi_hki'])
df_train['keterlibatan_vltr'] = encode.fit_transform(df_train['keterlibatan_vltr'])
df_train['jml_vltr_terlibat'] = encode.fit_transform(df_train['jml_vltr_terlibat'])
df_train['skema'] = encode.fit_transform(df_train['skema'])
df_train['luaran_hki'] = encode.fit_transform(df_train['luaran_hki'])
df_train['jenis_luaran_hki'] = encode.fit_transform(df_train['jenis_luaran_hki'])
df_train['luaran_lain'] = encode.fit_transform(df_train['luaran_lain'])
df_train['jenis_luaran_lain'] = encode.fit_transform(df_train['jenis_luaran_lain'])
df_train['status_luaran_lain'] = encode.fit_transform(df_train['status_luaran_lain'])

df_test['potensi_hki'] = encode.fit_transform(df_test['potensi_hki'])
df_test['keterlibatan_vltr'] = encode.fit_transform(df_test['keterlibatan_vltr'])
df_test['jml_vltr_terlibat'] = encode.fit_transform(df_test['jml_vltr_terlibat'])
df_test['skema'] = encode.fit_transform(df_test['skema'])
df_test['luaran_hki'] = encode.fit_transform(df_test['luaran_hki'])
df_test['jenis_luaran_hki'] = encode.fit_transform(df_test['jenis_luaran_hki'])
df_test['luaran_lain'] = encode.fit_transform(df_test['luaran_lain'])
df_test['jenis_luaran_lain'] = encode.fit_transform(df_test['jenis_luaran_lain'])
df_test['status_luaran_lain'] = encode.fit_transform(df_test['status_luaran_lain'])
```

```python
# Slice dataframe mulai dari kolom 'cap-shape' sampai 'habitat'
X_train = df_train.loc[:,['keterlibatan_vltr', 'jml_vltr_terlibat', 'skema', 'luaran_hki', 'jenis_luaran_hki', 'luaran_lain', 'jenis_luaran_lain', 'status_luaran_lain'] ]
y_train = df_train['potensi_hki']
X_test = df_test.loc[:,['keterlibatan_vltr', 'jml_vltr_terlibat', 'skema', 'luaran_hki', 'jenis_luaran_hki', 'luaran_lain', 'jenis_luaran_lain', 'status_luaran_lain'] ]
y_test = df_test['potensi_hki']
```

```python
dt = DecisionTreeClassifier(max_depth=6, random_state=1)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
acc_dt = accuracy_score(y_test, y_pred_dt)
```

```python
rf = RandomForestClassifier(n_estimators=3, random_state=5)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
acc_rf = accuracy_score(y_test, y_pred_rf)
```

```python
print("Hasil Decision Tree")
print("Test set accuracy: {:.2f}".format(acc_dt))
print(f"Hasil Detail: {acc_dt}")
print("Hasil Random Forest")
print("Test set accuracy: {:.2f}".format(acc_rf))
print(f"Hasil Detail: {acc_rf}")
```

# Soal No 5
Terapkan sebuah metode prediksi (berdasarkan tahun, misal: jumlah kegiatan pada tahun tertentu sebagai x dan jumlah  luaran sebagai y) serta hitung akurasinya dengan wajib melibatkan variabel keterlibatan volunteer sampai variabel luaran-luaran yang dihasilkan! (20 poin)

```python
df = data
prediksi = df[df['tahun'] == 2021]
prediksi_train, prediksi_test = train_test_split(prediksi, test_size=0.2, random_state=0)
print(f'Jumlah data asli: {prediksi.shape[0]}')
print(f'Jumlah data train: {prediksi_train.shape[0]}')
print(f'Jumlah data test: {prediksi_test.shape[0]}')
```

```python
X_train = prediksi_train.loc[:,['keterlibatan_vltr', 'jml_vltr_terlibat', 'skema', 'luaran_hki', 'jenis_luaran_hki', 'luaran_lain', 'jenis_luaran_lain', 'status_luaran_lain'] ]
y_train = prediksi_train['potensi_hki']
X_test = prediksi_test.loc[:,['keterlibatan_vltr', 'jml_vltr_terlibat', 'skema', 'luaran_hki', 'jenis_luaran_hki', 'luaran_lain', 'jenis_luaran_lain', 'status_luaran_lain'] ]
y_test = prediksi_test['potensi_hki']
dt = DecisionTreeClassifier(max_depth=6, random_state=1)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
acc_dt = accuracy_score(y_test, y_pred_dt)
print("Hasil Decision Tree")
print("Test set accuracy: {:.2f}".format(acc_dt))
print(f"Hasil Detail: {acc_dt}")
```
