##### 4. Terapkan dua jenis metode klasifikasi dan lakukan perbandingan akurasi  dengan wajib melibatkan variabel keterlibatan volunteer sampai variabel luaran-luaran yang dihasilkan! (20 poin)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# # Memisahkan label dan fitur
# X = df.iloc[:, 6:14] 
# y = df.iloc[:, 15] 

# X untuk kolom index ke-1 dst
X_train = df_train.iloc[:, 6:14]
X_test = df_test.iloc[:, 6:14] 
# y untuk kolom index ke-15
y_train = df_train.iloc[:, 15] 
y_test = df_test.iloc[:, 15]

# DT ENTROPY
dt_entropy = DecisionTreeClassifier(criterion='entropy')

# Fit Model
dt_entropy.fit(X_train, y_train)

# Memprediksi label set test
y_pred_entropy = dt_entropy.predict(X_test)

# Evaluasi
acc_entropy = accuracy_score(y_test, y_pred_entropy)

print(f'Akurasi DT Entropy: {acc_entropy}')

# RANDOM FOREST
rf = RandomForestClassifier(n_estimators=10, random_state=50)

# Fit Model
rf.fit(X_train, y_train)

# Memprediksi label set test
y_pred_rf = rf.predict(X_test)

# Evaluasi
acc_rf = accuracy_score(y_test, y_pred_rf)
print(f'Akurasi Random Forest: {acc_rf}')
```

Dari hasil akurasi tersebut, dapat dilihat bahwa model DT menggunakan entropy menghasilkan akurasi yang lebih baik dari pada model Random Forest

##### 5. Terapkan sebuah metode prediksi (berdasarkan tahun, misal: jumlah kegiatan pada tahun tertentu sebagai x dan jumlah  luaran sebagai y) serta hitung akurasinya dengan wajib melibatkan variabel keterlibatan volunteer sampai variabel luaran-luaran yang dihasilkan! (20 poin)

Membuat dataset baru berdasarkan dataset kuis untuk metode prediksi dalam menentukan jumlah status luaran (Draft dan Terpublikasi)

```python
df = pd.read_csv('dataset_quiz.csv')


dfp = pd.DataFrame(list())

dfp['tahun'] = df['tahun'].unique()

jml_tim = []
jml_vltr = []
reg = []
kem = []
jml_luaran_hki = []
jml_hak_cipta = []
ptn_ya = []
ptn_tdk = []
ptn_sdh = []
ptn_drf = []
jml_luaran_lain = []
jml_artikel = []
jml_media_massa = []
jml_modul_ajar = []
jml_status_draft_terpublikasi = []

for item in dfp['tahun']:
    slice = df[df.tahun == item]
    jml_tim.append(len(slice))
    jml_vltr.append(sum(slice['jml_vltr_terlibat']))
    reg.append(len(df[(df.tahun == item) & (df.skema == "REGULER")]))
    kem.append(len(df[(df.tahun == item) & (df.skema == "KEMITRAAN")]))
    jml_luaran_hki.append(len(df[(df.tahun == item) & (df.luaran_hki == "ADA")]))
    jml_hak_cipta.append(len(df[(df.tahun == item) & (df.jenis_luaran_hki== "Hak Cipta")]))
    ptn_ya.append(len(df[(df.tahun == item) & (df.potensi_hki == "YA")]))
    ptn_tdk.append(len(df[(df.tahun == item) & (df.potensi_hki == "TIDAK")]))
    ptn_sdh.append(len(df[(df.tahun == item) & (df.potensi_hki == "SUDAH")]))
    ptn_drf.append(len(df[(df.tahun == item) & (df.potensi_hki == "DRAFT")]))
    jml_luaran_lain.append(len(df[(df.tahun == item) & (df.luaran_lain == "YA")]))
    jml_artikel.append(len(df[(df.tahun == item) & (df.jenis_luaran_lain == "Artikel Ilmiah")]))
    jml_media_massa.append(len(df[(df.tahun == item) & (df.jenis_luaran_lain == "Publikasi Media Massa")]))
    jml_modul_ajar.append(len(df[(df.tahun == item) & (df.jenis_luaran_lain == "Modul Ajar")]))
    jml_status_draft_terpublikasi.append(len(slice[(slice.status_luaran_lain == "Terpublikasi") | (slice.status_luaran_lain == "Draft")]))


dfp['jml_tim'] = jml_tim
dfp['jml_vltr'] = jml_vltr
dfp['reg'] = reg
dfp['kem'] = kem
dfp['jml_luaran_hki'] = jml_luaran_hki
dfp['jml_hak_cipta'] = jml_hak_cipta
dfp['ptn_ya'] = ptn_ya
dfp['ptn_tdk'] = ptn_tdk
dfp['ptn_sdh'] = ptn_sdh
dfp['ptn_drf'] = ptn_drf
dfp['jml_luaran_lain'] = jml_luaran_lain
dfp['jml_artikel'] = jml_artikel
dfp['jml_media_massa'] = jml_media_massa
dfp['jml_modul_ajar'] = jml_modul_ajar
dfp['jml_status_draft_terpublikasi'] = jml_status_draft_terpublikasi 

dfp.to_csv('dataset_prediksi.csv')
dfp.head()
```

```python
fig, ax = plt.subplots(figsize=(30,20)) 
sns.heatmap(dfp.corr(), cmap='YlGnBu', annot = True, linewidths=.7, ax=ax)
plt.show()
```

```python
X = dfp[['jml_modul_ajar', 'jml_vltr', 'jml_luaran_hki', 'jml_modul_ajar', 'jml_artikel', 'jml_luaran_lain', 'ptn_drf', 'ptn_ya', 'ptn_sdh']]
y = dfp['jml_status_draft_terpublikasi']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

from sklearn.ensemble import RandomForestRegressor
import numpy as np
regressor = RandomForestRegressor(n_estimators = 1000, random_state = 42)
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

# Evaluasi
# Calculate the absolute errors
errors = abs(y_pred - y_test)

# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / y_test)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

##### 7. Lakukan hyperparameter tunning berdasarkan jawaban nomor 2, 3, dan 4 menggunakan metode Grid Search, Random Search, dan Exhausive Search! (20 poin)

```python
# Memisahkan label dan fitur
X = df.iloc[:, 6:14] # X untuk kolom index ke-1 dst
y = df.iloc[:, 15] # y untuk kolom index ke-15

from sklearn.model_selection import RepeatedKFold

# Inisiasi repated k-fold
rkf = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

# inisiasi model
dt = DecisionTreeClassifier()

# Definisikan hyperparameter yang akan digunakan
# sklearn menerima dalam bentuk dictionary
# nama hyperparamater HARUS SESUAI dengan dokumentasi sklearn
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5, 11)),
}

# inisiasi rs berdasarkan nilai repeated k-fold dan hyperparameter
rs = RandomizedSearchCV(dt, param_distributions=params,
                        n_iter=10, scoring="accuracy",
                        n_jobs=-1, cv=rkf, return_train_score=True)

# Fit / latih berdasarkan random
# %timeit merupakan magic command didalam ipython notebook
# yang dapat kita gunakan untuk menghitung waktu komputasi
# cara ini cukup efektif untuk melakukan evaluasi suatu algoritma atau prosedur
%timeit rs.fit(X, y)

# Evaluasi dengan score
score = rs.score(X, y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {rs.best_params_}')
```
