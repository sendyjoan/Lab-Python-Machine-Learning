Pembobotan TF-IDF menggunakan TfidVectorizer (Tugas pada Jobsheet 1)

```python
from numpy import asarray
import pandas as pd
import string

data = pd.read_csv
with open('/content/orange.txt', 'r') as data:
    text = data.read()
    words = text.split()
    table = str.maketrans("","", string.punctuation)
    stripped = [w.translate(table) for w in words]
    assembled = " ".join(stripped)
    corpus = [
        assembled
    ]

    from sklearn.feature_extraction.text import TfidfVectorizer

    vectorizer = TfidfVectorizer(stop_words='english')
    response = vectorizer.fit_transform(corpus)
    print(response)
```

<p> Pertama-tama kita lihat variable yang ada dengan meutilisasi method description</p>

<p>Method <b>describe</b> secara otomatis melakukan komputasi statistik untuk semua continous variable. Seluruh NaN akan ditidak dihitung didalam statistics.</p>

This will show:

<ul>
    <li>the count of that variable</li>
    <li>mean</li>
    <li>standard deviation (std)</li> 
    <li>nilai minimum</li>
    <li>IQR (Interquartile Range: 25%, 50% dan 75%)</li>
    <li>nilai maximum</li>
<ul>


 Kita dapat meutilisasi method "describe" seperti berikut ini:


```python
df.describe()
```

Secara default describe melakukan ignore terhada variabel bertype objek. Kita dapat menggunakan method describe pada variable object dengan cara sebagai berikut


```python
df.describe(include=['object'])
```


<h3>Value Counts</h3>
<p>Penghitungan nilai adalah cara yang baik untuk memahami berapa banyak unit dari setiap karakteristik/variabel yang kita miliki. Kita bisa menerapkan metode "value_counts" pada kolom 'album'. Jangan lupa metode "value_counts" hanya berfungsi pada seri Pandas, bukan Pandas Dataframes. Sehingga, kita hanya menyertakan satu tanda kurung "df['album']" bukan dua tanda kurung "df[['album']]".</p> 


```python
df['jiwa_terdampak'].value_counts()
```

Kita dapat melakukan konversi kedalam dataframe dengan cara sebagai berikut :


```python
df['jiwa_terdampak'].value_counts().to_frame()
```

Mari ulangi langkah di atas tetapi simpan hasilnya ke dataframe 'jiwa_terdampak' dan ganti nama kolom 'jiwa_terdampak' menjadi 'value_counts'. 


```python
jiwa_terdampak_counts = df['jiwa_terdampak'].value_counts().to_frame()
jiwa_terdampak_counts.rename(columns={'jiwa_terdampak': 'value_counts'}, inplace=True)
jiwa_terdampak_counts
```

```python
jiwa_terdampak_counts.index.name = 'jiwa_terdampak'
jiwa_terdampak_counts
```

Kita dapat mengulangi proses di atas untuk variabel 'kecamatan'. 


```python
# artist pop as variable
artist_pop_counts = df['kecamatan'].value_counts().to_frame()
artist_pop_counts.rename(columns={'kecamatan': 'value_counts'}, inplace=True)
artist_pop_counts.index.name = 'kecamatan'
artist_pop_counts.head(10)
```
