**Jawaban Tugas No 1**

```python
import numpy as np
import pandas as pd

df = pd.read_csv('https://storage.googleapis.com/sendyjoan_bucket1/wbc.csv')

# slice kolom terakhir --> Unnamed column
df = df.iloc[:,:-1]

df.head()
```

```python
X = df.iloc[:, 2:]
y = df['diagnosis']

X.head()
```

```python
from sklearn.model_selection import RepeatedKFold

# inisiasi repated k-fold
cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

# inisiasi model
dt = DecisionTreeClassifier()

# Definisikan hyperparameter yang akan digunakan
# sklearn menerima dalam bentuk dictionary
# nama hyperparamater HARUS SESUAI dengan dokumentasi sklearn
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': list(range(5,11))
}

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
search = RandomizedSearchCV(dt, params, cv=cv)

# Fit / latih berdasarkan grid
# %timeit merupakan magic command didalam ipython notebook
# yang dapat kita gunakan untuk menghitung waktu komputasi
# cara ini cukup efektif untuk melakukan evaluasi suatu algoritma atau prosedur
%timeit search.fit(X, y)

# Evaluasi dengan score
score = search.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {search.best_params_}')
```

**Jawaban Tugas 2**

GridSearch

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
# Regularization
# Regularization adalah suatu teknik untuk menghindari overfitting
# dengan cara menambahkan suatu nilai ke dalam fungsi cost function
# sehingga fungsi cost function yang dihasilkan tidak hanya memaksimalkan
# akurasi model, tetapi juga memaksimalkan nilai parameter yang digunakan
# dalam model
model = SVC()

params = {
    'kernel': ['linear', 'rbf', 'poly'],
    'degree': [2, 3, 4, 5],
    'C': [0.1, 1, 10, 100, 1000]
}

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid = GridSearchCV(model, param_grid=params, cv=cv)

# Fit / latih berdasarkan random
# %timeit merupakan magic command didalam ipython notebook
# yang dapat kita gunakan untuk menghitung waktu komputasi
# cara ini cukup efektif untuk melakukan evaluasi suatu algoritma atau prosedur

%timeit grid.fit(X, y)

# Evaluasi dengan score
score = grid.score(X, y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {grid.best_params_}')
```

Random Search

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
# Regularization
# Regularization adalah suatu teknik untuk menghindari overfitting
# dengan cara menambahkan suatu nilai ke dalam fungsi cost function
# sehingga fungsi cost function yang dihasilkan tidak hanya memaksimalkan
# akurasi model, tetapi juga memaksimalkan nilai parameter yang digunakan
# dalam model
model = SVC()

params = {
    'kernel': ['linear', 'rbf', 'poly'],
    'degree': [2, 3, 4, 5],
    'C': [0.1, 1, 10, 100, 1000]
}

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
random = RandomizedSearchCV(model, params, cv=cv)

# Fit / latih berdasarkan random
# %timeit merupakan magic command didalam ipython notebook
# yang dapat kita gunakan untuk menghitung waktu komputasi
# cara ini cukup efektif untuk melakukan evaluasi suatu algoritma atau prosedur

%timeit random.fit(X, y)

# Evaluasi dengan score
score = random.score(X, y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {random.best_params_}')
```

Manual

```python
# Setelah melakukan beberapa percobaan kombinasi hyperparameter,
# hasil evaluasi terbaik didapatkan dengan konfigurasi hyperparameter sebagai berikut
ms1 = SVC(kernel="linear", C=100)
ms2 = SVC(kernel="rbf", C=100)
ms3 = SVC(kernel="poly", C=100)

# Fit / latih berdasarkan random
# %timeit merupakan magic command didalam ipython notebook
# yang dapat kita gunakan untuk menghitung waktu komputasi
# cara ini cukup efektif untuk melakukan evaluasi suatu algoritma atau prosedur

%timeit ms1.fit(X, y)
%timeit ms2.fit(X, y)
%timeit ms3.fit(X, y)

# Evaluasi dengan score
score1 = ms1.score(X, y)
score2 = ms2.score(X, y)
score3 = ms3.score(X, y)

print(f'Hasil evaluasi kernel linear: {score1}')
print(f'Hasil evaluasi kernel rbf: {score2}')
print(f'Hasil evaluasi kernel poly: {score3}')

# Menggunakan hyperparameter gamma
msh1 = SVC(kernel="rbf", C=100, gamma=0.001)
msh2 = SVC(kernel="linear", C=100, gamma=0.001)
msh3 = SVC(kernel="poly", C=100, gamma=0.001)

%timeit msh1.fit(X, y)
%timeit msh2.fit(X, y)
%timeit msh3.fit(X, y)

score1 = msh1.score(X, y)
score2 = msh2.score(X, y)
score3 = msh3.score(X, y)

print(f'Hasil evaluasi kernel rbf dengan gamma: {score1}')
print(f'Hasil evaluasi kernel linear dengan gamma: {score2}')
print(f'Hasil evaluasi kernel poly dengan gamma: {score3}')
```
