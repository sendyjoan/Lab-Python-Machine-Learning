# *Melakukan Import Data dan Package*
Pada tahap ini saya melakukan import untuk package yang akan digunakan serta import file yang akan digunakan dalam operasi. Setelah dilakukan import dan printing head akan dilakukan pendeskripsian dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

# Load data CSV ke Data Frame
df = pd.read_csv('train.csv')

# Cek data
df.head()
```

# *Mencari Korelasi Data*
Pada tahap ini saya akan mencari korelasi data terdekat dari SalePrice yang mana saleprice akan diprediksi dan berupa outputan. Mencari korelasi ini menggunakan sns.heatmap

```python
fig, ax = plt.subplots(figsize=(20,20)) 
sns.heatmap(df.corr(), cmap="YlGnBu", annot = True)
plt.show()
```

Setelah dilakukan pencarian korelasi, maka diketahui bahwa ada 2 korelasi yang dapat diambil. yakni, OverallQual dan GrLivArea. Untuk selanjutnya akan dilihat melalui plot sebagai berikut

# *Melakukan Regresi Data*
Setelah diketahui kolom yang akan digunakan sebagai X dan Y. Maka selanjutnya akan dilakukan Regresi. Disini saya mengunakan GrLivArea

```python
#Regrasi
# Buat variabel bebas X dan Y, sebagai contoh ambil dari hasil analisis korelasi dari kegaitan sebelumnya
X = df['GrLivArea']
y = df['SalePrice']
```

Buat pemisahan data uji dan data latih dengan proporsi 7:3

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
```

Melihat hasil training dtaset

```python
X_train
y_train
```

Training model

```python
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
```

Fitting garis regresi

```python
lr = sm.OLS(y_train, X_train_sm).fit()
lr.params
```

Analisis statistika dari garis regresi

```python
lr.summary()
```

Visualisasi garis regresi

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, 2.335 + 103.5552 *X_train, 'r')
plt.show()
```

# *Residual Analysis*
Dipakai untuk mengetahui tingkat error dari variabel yang dipengaruhi (y)
Error = Actual y value - y predicted value

Prediksi y_value dari data x yang telah dilatih

```python
y_train_pred = lr.predict(X_train_sm)

res = (y_train - y_train_pred)
```

Cek histogram apakah berdistribusi normal atau tidak

```python
fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```

```python
plt.scatter(X_train,res)
plt.show()
```

```python
# prediksi pada data uji dan evaluasi model
X_test_sm = sm.add_constant(X_test)

# prediksi y value yang berkorelasi dengan X_test_sm
y_test_pred = lr.predict(X_test_sm)

# cetak 5 data terprediksi teratas
y_test_pred.head()
```

```python
# hitung nilai r^2
from sklearn.metrics import r2_score

r_squared = r2_score(y_test, y_test_pred)
r_squared
```

Prlu diperhatikan bahwa r^2 dari data trainign adalah 0.485 sedangkan dari data testing adalah 0.537. Hal ini berarti model yang dibentuk cukup stabil (tidak berselisih jauh antara training dengan testing)

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_test_pred, 'r')
plt.show()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=30)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Inisiasi obyek MultinomialNB
gnb = GaussianNB()

# Fit model
# Label y harus dalam bentu 1D atau (n_samples,)
gnb.fit(X_train, y_train)

# Prediksi dengan data training
y_train_pred = gnb.predict(X_train)

# Evaluasi akurasi training
acc_train = accuracy_score(y_train, y_train_pred)

# Prediksi test data
y_test_pred = gnb.predict(X_test)

# Evaluasi model dengan metric akurasi
acc_test = accuracy_score(y_test, y_test_pred)

# Print hasil evaluasi
print(f'Hasil akurasi data train: {acc_train}')
print(f'Hasil akurasi data test: {acc_test}')
```
